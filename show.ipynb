```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('data.csv', sep=';')
df
```

```python
scale = 3
fig, axs = plt.subplots(3,4, figsize=(4*scale,3*scale))

m='.'
l='--'

axs[0][0].plot(df["k"], df["bfs(n)"], marker=m, linestyle=l, c="blue")
axs[0][0].set_title("bfs(n)")
axs[0][1].plot(df["k"], df["bfs_mem(n)"], marker=m, linestyle=l, c="green")
axs[0][1].set_title("bfs_mem(n)")
axs[1][0].plot(df["k"], df["a*(n)"], marker=m, linestyle=l, c="red")
axs[1][0].set_title("a*(n)")
axs[1][1].plot(df["k"], df["a*_mem(n)"], marker=m, linestyle=l, c="orange")
axs[1][1].set_title("a*_mem(n)")

axs[0][2].plot(df["k"], df["bfs(t)"], marker=m, linestyle=l, c="blue")
axs[0][2].set_title("bfs(t)")
axs[0][3].plot(df["k"], df["bfs_mem(t)"], marker=m, linestyle=l, c="green")
axs[0][3].set_title("bfs_mem(t)")
axs[1][2].plot(df["k"], df["a*(t)"], marker=m, linestyle=l, c="red")
axs[1][2].set_title("a*(t)")
axs[1][3].plot(df["k"], df["a*_mem(t)"], marker=m, linestyle=l, c="orange")
axs[1][3].set_title("a*_mem(t)")

axs[2][0].plot(df["k"], df["bfs(n)"], marker=m, linestyle=l, c="blue")
axs[2][0].plot(df["k"], df["bfs_mem(n)"], marker=m, linestyle=l, c="green")
axs[2][0].plot(df["k"], df["a*(n)"], marker=m, linestyle=l, c="red")
axs[2][0].plot(df["k"], df["a*_mem(n)"], marker=m, linestyle=l, c="orange")
axs[2][0].set_title("bfs(n), bfs_mem(n),\n a*(n) and a*_mem(n)")

axs[2][1].plot(df["k"], df["bfs_mem(n)"], marker=m, linestyle=l, c="green")
axs[2][1].plot(df["k"], df["a*(n)"], marker=m, linestyle=l, c="red")
axs[2][1].plot(df["k"], df["a*_mem(n)"], marker=m, linestyle=l, c="orange")
axs[2][1].set_title("bfs_mem(n), a*(n)\n and a*_mem(n)")

axs[2][2].plot(df["k"], df["bfs(t)"], marker=m, linestyle=l, c="blue")
axs[2][2].plot(df["k"], df["bfs_mem(t)"], marker=m, linestyle=l, c="green")
axs[2][2].plot(df["k"], df["a*(t)"], marker=m, linestyle=l, c="red")
axs[2][2].plot(df["k"], df["a*_mem(t)"], marker=m, linestyle=l, c="orange")
axs[2][2].set_title("bfs(t), bfs_mem(t),\n a*(t) and a*_mem(t)")

axs[2][3].plot(df["k"], df["bfs_mem(t)"], marker=m, linestyle=l, c="green")
axs[2][3].plot(df["k"], df["a*(t)"], marker=m, linestyle=l, c="red")
axs[2][3].plot(df["k"], df["a*_mem(t)"], marker=m, linestyle=l, c="orange")
axs[2][3].set_title("bfs_mem(t), a*(t)\n and a*_mem(t)")

fig.tight_layout()
plt.show()
```
